## Scrape Well being

```python
from bs4 import BeautifulSoup
import requests
```

```python
link = r'https://facileimmobiliare.it/blog/classifica-qualita-della-vita-2022/'
page = requests.get(link)

soup1 = BeautifulSoup(page.content, 'html.parser')
soup2 = BeautifulSoup(soup1.prettify(), 'html.parser')
rows = soup2.find('tbody').find_all('tr')
rows
```

```python
len(rows)
```

```python
rows = rows[1:]
```

```python
province = []
score = []

for r in rows:
    r2 = r.find_all('td')
    province.append(r2[1].get_text(strip=True))
    score.append(r2[2].get_text(strip=True))
```

```python
print(len(province), len(score))
```

```python
data = {'province':province, 'score':score}
df_wb = pd.DataFrame(data)
df_wb
```

### Provinces scraping

```python
url = 'https://it.wikipedia.org/wiki/Province_d%27Italia'

page = requests.get(url)
```

```python
soup1 = BeautifulSoup(page.content, 'html.parser')
soup2 = BeautifulSoup(soup1.prettify(), 'html.parser')

table = soup2.find('tbody')
table
```

```python
rows = table.find_all('tr')
rows = rows[1:-1]
```

```python
len(rows)
```

```python
province2 = []
region2 = []
pop2 = []

for r in rows:
    r2 = r.find_all('td')
    province2.append(r2[0].get_text(strip=True))
    region2.append(r2[2].get_text(strip=True))
    pop2.append(r2[3].get_text(strip=True))
```

```python
print(len(region2),len(province2), len(pop2))
```

```python
data2 = {'province':province2, 'region':region2,'pop':pop2}

df_pop = pd.DataFrame(data2)
df_pop
```

```python
cleaning = df_pop['province'].str.find('[')
cleaning
```

```python
cleaning = cleaning.replace(-1, np.nan)
cleaning
```

```python
df_pop['province'] = df_pop['province'].str.extract(r'([^\[]*)')
df_pop['province']
```
